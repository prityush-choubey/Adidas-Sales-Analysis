```python
df["Invoice Date"]=pd.to_datetime(df["Invoice Date"])
```

```python
df=df.drop_duplicates()
```

```python
df["year"]=df["Invoice Date"].dt.year
df["month"]=df["Invoice Date"].dt.month
df["month_name"]=df["Invoice Date"].dt.month_name()
```

```python
df["revenue_per_unit"]=df["Total Sales"]/df["Units Sold"]
```

```python
df["profit_per_unit"]=df["Operating Profit"]/df["Units Sold"]
```

```python
df.groupby ("Sales Method") ["Total Sales"].mean()
```

```python
df.groupby("Sales Method")["Total Sales"].sum().sort_values(ascending=False)
```

```python
plt.figure()
sns.boxplot(x=df["Units Sold"])
plt.title("units sold distribution")
plt.show
```

```python
plt.figure()
sns.scatterplot (data=df, x='Units Sold', y='Total Sales', hue='Sales Method')
plt.title("revenue vs units sold")
plt.show()
```
